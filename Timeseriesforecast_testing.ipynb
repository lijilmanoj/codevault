```python
import pandas as pd
import numpy as np

# Set the seed for reproducibility
np.random.seed(42)

# Define the number of rows and the date range
n_rows = 1000
start_date = '2021-01-01'
end_date = '2023-12-31'

# Generate the date column
dates = pd.date_range(start=start_date, end=end_date, periods=n_rows)

# Define the region and product hierarchies
region_levels = ['North', 'South', 'East', 'West']
product_levels = ['Electronics', 'Fashion', 'Home Goods', 'Toys']

region_level1 = np.random.choice(region_levels, size=n_rows)
region_level2 = np.random.choice(['Urban', 'Rural'], size=n_rows)
region_level3 = np.random.choice(['City', 'Town', 'Village'], size=n_rows)

product_level1 = np.random.choice(product_levels, size=n_rows)
product_level2 = np.random.choice(['Premium', 'Standard', 'Basic'], size=n_rows)
product_level3 = np.random.choice(['New', 'Used', 'Refurbished'], size=n_rows)

# Generate the revenue column
revenue = np.random.randint(100, 10000, size=n_rows)

# Create the dataset
data = pd.DataFrame({
    'date': dates,
    'region_level1': region_level1,
    'region_level2': region_level2,
    'region_level3': region_level3,
    'product_level1': product_level1,
    'product_level2': product_level2,
    'product_level3': product_level3,
    'revenue': revenue
})

# Save the dataset to a CSV file
#data.to_csv('data.csv', index=False)
```
